```python
import duckdb
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
df = con.execute("select * from produtos").fetch_df()
df.head()
```

```python
df2 = df.drop(columns=['categoria', 'fornecedor'])
df2.head()
```

```python
con.execute("drop table dim_produtos") 
```

```python
con.execute("""
create table if not exists dim_produtos (
            id BIGINT,
            nome TEXT,
            preco FLOAT
            )
""")
```

```python
con.execute("insert into dim_produtos select * from df2 order by id")
```

```python
df_dim = con.execute("select * from dim_produtos").fetch_df()
df_dim.head()
```
